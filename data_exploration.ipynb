```python
import pandas
data = pandas.read_csv("train.csv")
```

```python
print data.head()
print data.describe()
```

```python
data.isnull().sum()
```

Seems like there are a lot of missing data in Age and Cabin

```python
#grouped by sex
data.groupby("Sex").mean()
# female's survival rate is almost 4 times 
```

```python
data.groupby("Pclass").mean()
# the high seat had a higher survival rate than the lower one even though average age is greater
```

```python
young_male = data.Survived[data.Age <= 16][data.Sex == "male"]
young_female = data.Survived[data.Age <= 16][data.Sex == "female"]
adult_male = data.Survived[data.Age > 16][data.Sex == "male"]
adult_female = data.Survived[data.Age > 16][data.Sex == "female"]
young = data.Survived[data.Age <= 16]
adult = data.Survived[data.Age > 16]

print "Survival rate by age"
print
print "young", young.mean(), young.count()
print "adult", adult.mean(), adult.count()
print "young male", young_male.mean(), young_male.count()
print "young female", young_female.mean(), young_female.count()
print "adult male", adult_male.mean(), adult_male.count()
print "adult female", adult_female.mean(), adult_female.count()
```

```python
data.groupby("Fare").mean()
```

```python
poor = data.Survived[data.Fare <= 10]
normal = data.Survived[(data.Fare > 10) & (data.Fare <= 20)]
rich = data.Survived[(data.Fare >20) & (data.Fare <= 50)]
richer = data.Survived[(data.Fare >50)]

print "Survival rate by Money"
print
print "under 10", poor.mean(), poor.count()
print "10 ~ 20", normal.mean(), normal.count()
print "20 ~ 50", rich.mean(), rich.count()
print "over 50",richer.mean(), richer.count()
```

People who paid more for the ticket (rich people) had higher survival rate 

```python
poor_female = data.Survived[data.Fare <= 10][data.Sex == "female"]
poor_male = data.Survived[data.Fare <= 10][data.Sex == "male"]
normal_female = data.Survived[(data.Fare > 10) & (data.Fare <= 20)][data.Sex == "female"]
normal_male = data.Survived[(data.Fare > 10) & (data.Fare <= 20)][data.Sex == "male"]
rich_female = data.Survived[(data.Fare >20) & (data.Fare <= 50)][data.Sex == "female"]
rich_male = data.Survived[(data.Fare >20) & (data.Fare <= 50)][data.Sex == "male"]
richer_female = data.Survived[(data.Fare >50)][data.Sex == "female"]
richer_male = data.Survived[(data.Fare >50)][data.Sex == "male"]

print "Survival rate by gender and money"
print 
print "poor female", poor_female.mean(), poor_female.count()
print "poor male", poor_male.mean(), poor_male.count()
print "poor female", normal_female.mean(), normal_female.count()
print "poor male", normal_male.mean(), normal_male.count()
print "poor female", rich_female.mean(), rich_female.count()
print "poor male", rich_male.mean(), rich_male.count()
print "poor female", richer_female.mean(), richer_female.count()
print "poor male", richer_male.mean(), richer_male.count()
```

Females definitely have a higher survival rate and the survival rate also goes up with people's socio-economic status assuming the Fares they paid is correlated to their socio-economic status
